```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('netflix_titles.csv')
```

```python
df_genres = df.copy()
df_genres['listed_in'] = df_genres['listed_in'].str.split(', ')
df_exploded_genres = df_genres.explode('listed_in')
df_recent_genres = df_exploded_genres[df_exploded_genres['release_year'] >= 2010]
top_10_genres = df_recent_genres['listed_in'].value_counts().nlargest(10).index
df_top_genres = df_recent_genres[df_recent_genres['listed_in'].isin(top_10_genres)]
trend_data = df_top_genres.groupby(['release_year', 'listed_in']).size().unstack().fillna(0)
```

```python
plt.figure(figsize=(12, 7))
sns.lineplot(data=trend_data, dashes=False, marker='o')
plt.title('Tren Popularitas Genre di Netflix (2010 - 2021)', fontsize=15)
plt.xlabel('Tahun Rilis', fontsize=12)
plt.ylabel('Jumlah Konten', fontsize=12)
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('genre_trend_netflix.png') # Menyimpan grafik 1
plt.show()
```

```python
df_country = df.dropna(subset=['country']).copy()
df_country['country'] = df_country['country'].str.split(', ')
df_exploded_country = df_country.explode('country')
top_10_countries_names = df_exploded_country['country'].value_counts().nlargest(10).index
df_top_countries = df_exploded_country[df_exploded_country['country'].isin(top_10_countries_names)]
```

```python
plt.figure(figsize=(10, 8))
sns.countplot(data=df_top_countries, y='country', hue='type',
              order=top_10_countries_names, palette='pastel')
plt.title('Top 10 Negara Kontributor Konten Netflix', fontsize=15)
plt.xlabel('Jumlah Konten', fontsize=12)
plt.ylabel('Negara', fontsize=12)
plt.legend(title='Tipe Konten')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('country_distribution_netflix.png') # Menyimpan grafik 2
plt.show()
```
